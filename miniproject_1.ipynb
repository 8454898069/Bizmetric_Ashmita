```python
cart = []

try:
    a = input("Do you want Books(1) or NoteBook(2): ").strip()

    if a == '1':
        try:
            standard = int(input("Enter Standard: "))
        except ValueError:
            print("Invalid standard! Please enter a number.")
            exit()

        if 1 <= standard <= 4:
            n_standard = '1st-4th'
        elif 5 <= standard <= 8:
            n_standard = '5th-8th'
        elif 9 <= standard <= 10:
            n_standard = '9th-10th'
        else:
            print("Standard out of range!")
            exit()

        my_dict = {
            '1st-4th': {'Books': {'hindi': 60, 'marathi': 60, 'english': 80, 'science': 90, 'maths': 100}},
            '5th-8th': {'Books': {'hindi': 100, 'marathi': 100, 'english': 100, 'science': 120, 'maths': 140}},
            '9th-10th': {'Books': {'hindi': 150, 'marathi': 150, 'english': 150, 'science': 200, 'maths': 250}}
        }

        try:
            n = int(input("How many books do you want? "))
        except ValueError:
            print("Invalid number")
            exit()

        for i in range(n):
            try:
                subject = input("Enter Book Name: ").lower().strip()
                qty = int(input("Enter Quantity: "))

                price = my_dict[n_standard]['Books'][subject]
                cart.append([subject, price, qty, price * qty])

            except KeyError:
                print("Book not found ")
            except ValueError:
                print("Invalid quantity")

    elif a == '2':
        try:
            pages = int(input("Enter Pages (100/200): "))
        except ValueError:
            print("Invalid pages")
            exit()

        if pages == 100:
            n_pages = 'Pages100'
        elif pages == 200:
            n_pages = 'Pages200'
        else:
            print("Only 100 or 200 pages allowed")
            exit()

        my_dict2 = {
            'Pages100': {'square': 40, '4lines': 30, '2lines': 20, 'single lines': 60, 'a4 notebook': 100},
            'Pages200': {'square': 70, '4lines': 50, '2lines': 50, 'single lines': 100, 'a4 notebook': 180}
        }

        try:
            n = int(input("How many notebooks do you want? "))
        except ValueError:
            print("Invalid number!")
            exit()

        for i in range(n):
            try:
                book_type = input("Enter Notebook Type: ").lower().strip()
                qty = int(input("Enter Quantity: "))

                price = my_dict2[n_pages][book_type]
                cart.append([book_type, price, qty, price * qty])

            except KeyError:
                print("Notebook not found")
            except ValueError:
                print("Invalid quantity")

    else:
        print("Invalid choice-- Please select 1 or 2.")
        exit()

    bill = input("Do you want bill? (yes/no): ").lower().strip()

    if bill == 'yes':
        print("----------- BILL -----------")
        print("Sr.No\tName\t\tPrice\tQty\tTotal")

        grand_total = 0
        for i, item in enumerate(cart, start=1):
            print(f"{i}\t{item[0]}\t\t{item[1]}\t{item[2]}\t{item[3]}")
            grand_total += item[3]

        print("-----------------------------")
        print(f"Grand Total: â‚¹{grand_total}")
        print("-----------------------------")

except Exception as e:
    print("Unexpected error:", e)
```
