```python
funct1 = lambda a, b: a + b
print(funct1(5, 3))
```

```python
funct2 = lambda x : x % 2 == 0
print(funct2(4))
```

```python
funct3 = lambda s: s[-1]
print(funct3("asmitha"))
```

```python
nums = [1,2,3,4]
funct4 = list(map(lambda x: x**2, nums))
print(funct4)
```

```python
nums = [1,2,3,4,5]
funct5 = list(filter(lambda x: x % 2 != 0, nums))
print(funct5)
```

```python
funct6 = lambda s: s == s[::-1]
print(funct6("sds"))
```
